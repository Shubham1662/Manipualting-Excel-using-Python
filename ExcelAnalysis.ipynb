Import the libraries needed 


```python
import pandas as pd #Here i only need pandas library
```

Import all the excel files and related sheets 

```python
orders=pd.read_excel('Global Superstore.xls',sheet_name='Orders') # First importing order sheet
```

```python
orders.head(2)
```

```python
orders=orders.drop(["Customer ID","Customer Name", "Order Priority","Sales","Quantity","Discount","Profit","Product Name","Product ID"],axis=1)
# deleted columns i don't need
```

```python
orders.head(2)
```

```python
return_sheet = pd.read_excel('Global Superstore.xls',sheet_name='Return') # Importing return sheet from same excel file
```

```python
return_sheet.head(2) # Courier aount paid by customer while returing product
```

```python
result=orders.merge(return_sheet,on='Order ID') # Merging both the sheet like Vllokup
```
