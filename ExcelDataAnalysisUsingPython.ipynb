```python
#Now from column Country, we have to remove all data that have Italy
Orders=Orders[~(Orders['Country']=='Italy')]
```

```python
#Now deleting all rows that have "CA" contains in Order ID like "CA-2012-124891"
Orders=Orders[~Orders['Order ID'].str.contains('CA',na=False)]
```

```python
#Now we will split Order ID into 3 parts on basis of "-" SA-2011-1830
Orders[['ID_Country','ID_Year','ID_Orders']]=Orders['Order ID'].str.split("-",expand=True)
```

```python
list(Orders.columns) #to Check headers name 
```

```python
#To delete a column
Orders=Orders.drop('City',axis=1)
```

```python
list(Orders.columns)#City column is deleted.
```

```python
#If we have to replace values in a column
#we will replace Customer Name Justin Ritter with Matt De Blank
#Let's first check cutomer name column
Orders['Customer Name'].head(4)
```

```python
Orders['Customer Name']=Orders['Customer Name'].replace("Justin Ritter","Matt De Blank")
```

# Time to Use Look up Function of Excel

```python
#For look of excel or fteching data from other excel Sheet we use Merge function in python.
#I will show with an example
```

```python
#Let's check our Return data
Returns.head(6)
```

```python
#For these Order IDs we will use our Orders dataset aand will fetch all information available
Returns=pd.merge(Returns,Orders,on='Order ID',left_index=True,right_index=False,how='left')
```

```python
#Now lets check Returns
Returns.head(3)
```

```python
#We can also sort Values in a column like A to Z or Newest to Oldest for date.
Orders=Orders.sort_values(["Order Date"],axis=0,ascending=False)
```

# Creating a Pivot table

```python
#Using Python we can also create Pivot Table in our excel ile
Pv1=Orders.pivot_table(values='Order ID',index=["Market","Ship Mode"],columns='Segment',margins=True,aggfunc='count')
```
